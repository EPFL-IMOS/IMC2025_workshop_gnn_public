```python
# %%
import torch
from torch import nn
from torch import Tensor
from torch.utils.data import Dataset

from torch_geometric.loader import DataLoader as GeometricDataLoader
from torch_geometric.data import Data
import torch_geometric.utils as tgu
import torch_geometric.nn as geom_nn

import lightning as pl
from lightning.pytorch import loggers as pl_loggers
import torchmetrics

import numpy as np
import networkx as nx
import matplotlib.pyplot as plt


def most_simple_graph(i):

  graph_1 = [
    [0,1],
    [0,3],
    [1,3],
    [1,2],
    [3,4],
    [3,2]
  ]

  graph_2 = [
    [0,1],
    [0,3],
    [1,3],
    [2,2],
    # [2,1]
  ]


  graph_3 = [
    [0,1],
    [1,2],
    [3,2],
    [3,4],
    [4,0],
    [4,2]
  ]

  graph_index = {
    0: graph_1,
    1: graph_2,
    2: graph_3,
  }

  ei = np.array(graph_index[i])
  # Converting undirected graph to directed graph implies adding the reverse edge
  G = nx.from_edgelist(ei).to_directed()
  edge_tensor = torch.tensor(list(G.edges), dtype=torch.long)
  return edge_tensor
```
